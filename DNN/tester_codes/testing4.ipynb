```python
import tensorflow as tf
import numpy as np
```

```python
x = []
for i in range(1000):
    x.append(i)

y = []
for i in range(1000):
    y.append(i)
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1))

for i in range(1,5):
    model.add(tf.keras.layers.Dense(50, activation='relu'))
model.add(tf.keras.layers.Dense(1))
```

```python
model.compile(optimizer='adam',
                loss=tf.keras.losses.MeanSquaredError(),
                metrics=['accuracy'])
```

```python
x_train = np.reshape(x, newshape=(1000,1))
y_train = np.array(y)
```

```python
model.fit(x_train, 
                y_train,  
                epochs=500,
                batch_size=50,
                shuffle=True)
```
